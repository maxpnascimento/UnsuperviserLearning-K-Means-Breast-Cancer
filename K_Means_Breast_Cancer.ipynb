Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sb
from sklearn.cluster import KMeans 
```

Data

```python
#dataset 
df = pd.read_csv("wdbc.csv", header=None)
df
```

```python
#sb.pairplot(df,hue=0)                  #Data scatter view with labels        
#sb.pairplot(df)                        #Data scatter view without labels 

data = np.array(df.drop(0,axis = 1))    #Removing the Labels Column
data
```

Model

```python
kmeans = KMeans(n_clusters=2, max_iter= 400, random_state=0)   #K-Means: 2 Clusteres, 400 iteration
```

Train

```python
kmeans.fit(data)                              #Training
```

Test

```python
                                       
df['K-Means Classification'] = kmeans.labels_    #Inserindo uma coluna com os agrupamentos K-Means
df
```

```python
contVP, contVN, contFP, contFN, cont = 0, 0, 0, 0, 0

for i in np.array(df):
  if i[0]==i[31]:
    if (i[0] == 1) and (i[31]==1):
      contVP += 1
    elif (i[0] == 0) and (i[31]==0):
      contVN += 1
    cont += 1
  else:
    if (i[0] == 0) and (i[31]==1):
      contFP += 1
    elif (i[0] == 1) and (i[31]==0):
      contFN += 1

print("Acurácia: ", cont/len(np.array(df)), "\nVP: ", contVP, "\nVN: ", contVN, "\nFP: ", contFP, "\nFN: ", contFN)
```
